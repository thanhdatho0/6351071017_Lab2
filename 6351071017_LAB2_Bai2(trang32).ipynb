Đề bài: Một mô hình AI được dùng để phát hiện email spam. Sau khi chạy trên 100 email, kết quả thu được như sau:

    Phát hiện đúng 30 email là spam (True Positive - TP).

    Phát hiện nhầm 5 email thường thành spam (False Positive - FP).

    Bỏ sót 10 email spam (dự đoán là email thường) (False Negative - FN).

Yêu cầu: Hãy tính giá trị F-measure của mô hình.

```python
print("Hồ Thành Đạt - 6351071017")

from sklearn.metrics import precision_score, recall_score, f1_score
# Giả sử y_true là nhãn thực tế và y_pred là nhãn mô hình dự đoán
# Positive là 1, Negative là 0

# Dựa trên ví dụ:
# 30 TP: 30 mẫu có nhãn 1 và dự đoán là 1
# 5 FP: 5 mẫu có nhãn 0 nhưng dự đoán là 1
# 10 FN: 10 mẫu có nhãn 1 nhưng dự đoán là 0
y_true = [1]*40 + [0]*60 # 40 email spam, 60 email thường
y_pred = [1]*35 + [0]*5 + [0]*10 + [0]*50 # Mô hình dự đoán 35 cái là spam, trong đó có 5 cái sai (FP)

# Tính toán
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print(f"Giá trị Precision: {precision:.3f}")
print(f"Giá trị Recall: {recall:.3f}")
print(f"Giá trị F1-score (F-measure): {f1:.3f}")
```
