```python
STUDENT = "Đinh Quốc Bảo - 6351071006"
```

# Câu 4

```python
import pandas as pd
from sklearn.model_selection import train_test_split
print(STUDENT)
df = pd.read_csv('mushrooms.csv')

features = df.drop('class', axis=1)
labels = df['class']
```

# Câu 5

```python
print(STUDENT)
features.select_dtypes(exclude=['int64']).columns
```

```python
print(STUDENT)
features_onehot = pd.get_dummies(features, columns=features.select_dtypes(exclude=['int64']).columns)
features_onehot
```

# Câu 6

```python
print(STUDENT)
X_train, X_test, y_train, y_test = train_test_split(
    features_onehot, labels, test_size=0.3, random_state=42
)
```

# Câu 7

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
print(STUDENT)
id3 = DecisionTreeClassifier(criterion='entropy', random_state=42)
id3.fit(X_train, y_train)
```

```python
print(STUDENT)
y_pred = id3.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Report:", classification_report(y_test, y_pred, digits=2), sep='\n')
```

```python
print(STUDENT)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['e', 'p'], yticklabels=['e', 'p'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(STUDENT)
plt.figure(figsize=(20,12))
tree.plot_tree(id3, feature_names=X_train.columns, class_names=['e','p'],
               filled=True, rounded=True, fontsize=9)
plt.savefig('decision_tree.png', dpi=100, bbox_inches='tight')
plt.show()
```

# Câu 8

```python
print(STUDENT)
id3_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
id3_gini.fit(X_train, y_train)
```

```python
print(STUDENT)
y_pred_gini = id3_gini.predict(X_test)

print("Accuracy (Gini):", accuracy_score(y_test, y_pred_gini))
print("Report (Gini):", classification_report(y_test, y_pred_gini, digits=2), sep='\n')
```

```python
print(STUDENT)
cm_gini = confusion_matrix(y_test, y_pred_gini)
plt.figure(figsize=(6,5))
sns.heatmap(cm_gini, annot=True, fmt='d', cmap='Blues',
            xticklabels=['e', 'p'], yticklabels=['e', 'p'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Gini)')
plt.show()
```

```python
print(STUDENT)
plt.figure(figsize=(20,12))
tree.plot_tree(id3_gini, feature_names=X_train.columns, class_names=['e','p'],
               filled=True, rounded=True, fontsize=9)
plt.savefig('decision_tree_gini.png', dpi=100, bbox_inches='tight')
plt.show()
```
